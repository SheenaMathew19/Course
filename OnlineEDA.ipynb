```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
zomato_df = pd.read_csv("zomato.csv",encoding="latin-1")
zomato_df.head()
```

```python
country_code_df=pd.read_excel("Country_Code.xlsx")
country_code_df.head()
```

```python
zomato_df = zomato_df.merge(country_code_df[["Country Code","Country"]], on="Country Code",how="left")
zomato_df.head()
```

```python
zomato_df.columns
```

```python
zomato_df["Country"]=zomato_df["Country_x"] if "Country_x" in zomato_df.columns else zomato_df["Country"]
zomato_df["Country"]=zomato_df["Country_y"] if "Country_y" in zomato_df.columns else zomato_df["Country"]
zomato_df=zomato_df.drop("Country_x",axis=1,errors="ignore")
zomato_df=zomato_df.drop("Country_y",axis=1,errors="ignore")
zomato_df.head()
```

```python
zomato_df.columns
```

```python
filtered_ratings=zomato_df[zomato_df["Aggregate rating"]>0]
plt.figure(figsize=(8,4), dpi=100)
sns.histplot(filtered_ratings['Aggregate rating'],bins=20,kde=True,color='Blue')
plt.xlabel('Aggregate rating')
plt.ylabel('Frequency')
plt.title('Distribution of Aggregate rating')
plt.show()
```

 What is the average price range of restaurants across different cities? List the cities with highest range and the ones with lowest range. Draw conclusions from the output.

(Note that here price range values are integer values from 1 to 4, 1 representing an affordable range of restaurants, and 4 would indicate expensive restaurants).  

To answer this question, you can calculate the mean price range for each city.


```python
# 1. Group by City and calculate mean Price range, sorted descending
city_price_rangemean=zomato_df.groupby("City")["Price range"].mean().sort_values(ascending=False)


# 2. Get highest and lowest average price range
highest_price=city_price_rangemean.max()
lowest_price=city_price_rangemean.min()


# 3. List all cities with the highest average price range
highest_cities = city_price_rangemean[city_price_rangemean == highest_price].index.tolist()

# 4. List all cities with the lowest average price range
lowest_cities = city_price_rangemean[city_price_rangemean == lowest_price].index.tolist()

# 5. Print the results
print("Average Price Range per City (Descending):\n", city_price_rangemean, "\n")
print(f"Highest Average Price Range: {highest_price:.2f}")
print("Cities with Highest Average Price Range:", highest_cities, "\n")
print(f"Lowest Average Price Range: {lowest_price:.2f}")
print("Cities with Lowest Average Price Range:", lowest_cities)
```

5. Does the availability of the feature 'Has Table booking' impact the aggregate rating? Draw inferences from the output.

To create the boxplot excluding the 0 ratings, we'll first filter the dataset to remove restaurants with a rating of 0. Then, we will generate the boxplot to compare the aggregate ratings between restaurants with and without table booking.

```python
#Filter out the restaurants with 0 ratings
valid_ratings=zomato_df[zomato_df["Aggregate rating"]>0]
valid_ratings

# Boxplot for 'Has Table booking' vs 'Aggregate rating' (excluding 0 ratings)
plt.figure(figsize=(8,4), dpi=100)
sns.boxplot(x='Has Table booking', y='Aggregate rating', data=valid_ratings, palette="viridis")
plt.xlabel('Has Table Booking')
plt.ylabel('Aggregate Rating')
plt.title('Aggregate Rating vs Has Table Booking (Excluding 0 Ratings)')
plt.show()


# Calculate the average ratings for restaurants with and without table booking (excluding 0 ratings)
avg_ratings=valid_ratings.groupby("Has Table booking")["Aggregate rating"].mean()
print(avg_ratings)

# Display the average ratings
```

```python
zomato_df['Tab-booking_numerical'] = zomato_df['Has Table booking'].map({'Yes': 1, 'No': 0})

# Filter valid ratings
valid_ratings = zomato_df[zomato_df['Aggregate rating'] > 0]

# Correlation with Aggregate rating
correlation = valid_ratings['Tab-booking_numerical'].corr(valid_ratings['Aggregate rating'])
print("Pearson correlation:", correlation)
```

6. Does having an online delivery option influence the restaurant's rating? Which country's restaurant has the best and which one has the worst rating out of those who do not offer online delivery? Draw inferences from the output.

To investigate whether having an online delivery option influences restaurant ratings, we can compare the ratings of restaurants that offer online delivery and those that do not. We will use a box plot to visually compare the distributions of ratings for both groups (We are not including restaurants that have a 0 rating).

```python
# Filter out the restaurants with a 0 rating
valid_ratings = zomato_df[zomato_df['Aggregate rating'] > 0]

# 2. Boxplot: Has Online delivery vs Aggregate rating
plt.figure(figsize=(8, 6))
plt.boxplot(
    [valid_ratings[valid_ratings['Has Online delivery'] == 'Yes']['Aggregate rating'],
     valid_ratings[valid_ratings['Has Online delivery'] == 'No']['Aggregate rating']],
    labels=['Online Delivery', 'No Online Delivery']
)

# 3. Ratings by country for restaurants WITHOUT online delivery
no_delivery = valid_ratings[valid_ratings['Has Online delivery'] == 'No']
country_ratings = no_delivery.groupby('Country')['Aggregate rating'].mean().sort_values(ascending=False)

# 4. Best and worst country
best_country = country_ratings.idxmax()
worst_country = country_ratings.idxmin()

best_rating = country_ratings.max()
worst_rating = country_ratings.min()

print("Best country without online delivery:", best_country, "with avg rating:", best_rating)
print("Worst country without online delivery:", worst_country, "with avg rating:", worst_rating)
```

 What is the distribution of restaurants with and without table booking option, that serve only 'Italian', only 'Chinese' and only 'American' cuisines? Draw inferences from the output.

To analyze the distribution of restaurants that serve only 'Italian', 'Chinese', and 'American' cuisines, and whether they offer table booking, we can follow these steps:

```python
# Filter for restaurants that serve ONLY Italian, Chinese, or American cuisines
filtered_restaurants = zomato_df[zomato_df['Cuisines'].isin(['Italian', 'Chinese', 'American'])]

# Check the first few rows
print(filtered_restaurants[['Restaurant Name', 'Cuisines', 'Has Table booking']].head())
plt.figure(figsize=(8,6))
sns.countplot(data=filtered_restaurants, x='Cuisines', hue='Has Table booking', palette='Set2')

# Add title and labels
plt.title("Distribution of Table Booking Option for Italian, Chinese, and American Restaurants")
plt.xlabel("Cuisine")
plt.ylabel("Number of Restaurants")
plt.legend(title='Has Table Booking')
plt.show()
```

Since 90% of the restaurants listed on Zomato in this data are from India, find out if the average cost for two has an influence on ratings in India. Draw inferences from the output.

To explore whether there is a relationship between the average cost for two and the restaurant ratings (Aggregate rating), we can calculate the correlation coefficient between the two variables and visualize their relationship using a scatter plot.

```python
india_df = zomato_df[zomato_df['Country'] == 'India']
india_df = india_df[(india_df['Aggregate rating'] > 0) & (india_df['Average Cost for two'] > 0)]
cost_rating_df = india_df[['Average Cost for two', 'Aggregate rating']]
plt.figure(figsize=(8,6))
sns.scatterplot(data=cost_rating_df, x='Average Cost for two', y='Aggregate rating')
plt.title('Average Cost for Two vs Aggregate Rating (India)')
plt.xlabel('Average Cost for Two')
plt.ylabel('Aggregate Rating')
plt.show()
correlation = cost_rating_df['Average Cost for two'].corr(cost_rating_df['Aggregate rating'])
print("Correlation between Average Cost for Two and Aggregate Rating:", correlation)
```

0. How does the price range affect the average ratings of restaurants, and how is this influenced by the presence of online delivery? Draw inferences from the output.

```python
# Group by Price range and Online delivery, then calculate mean rating
rating_by_price_delivery = zomato_df.groupby(['Price range', 'Has Online delivery'])['Aggregate rating'].mean().reset_index()
print(rating_by_price_delivery)


plt.figure(figsize=(8,6))
sns.lineplot(data=rating_by_price_delivery, x='Price range', y='Aggregate rating', hue='Has Online delivery', marker='o')

# Add title and labels
plt.title('Average Rating by Price Range and Online Delivery')
plt.xlabel('Price Range')
plt.ylabel('Average Aggregate Rating')
plt.xticks([1,2,3,4])  # Assuming price range is 1 to 4
plt.show()
```
