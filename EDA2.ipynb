```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#read the data set of "bank telemarketing campaign" in inp0.
df= pd.read_csv("Employee_Specifics.csv",index_col=0)
df.head()
```

```python
filtered_employees = df[(df['Rating'] == 5) & (df['Compensation'] < 700000)]
```

```python
employee_names = filtered_employees['Name'].tolist()

# What this does:
# - filtered_employees['Employee_Name'] → Gets the Employee_Name column as pandas Series
# - .tolist()                          → Converts pandas Series to Python list

# STEP 3: Enumerate and Display Names
for i, name in enumerate(employee_names, 1):
    print(f"{i}. {name}")
```

```python
plt.figure(figsize=(8,4), dpi=100)
sns.scatterplot(data=df, x="Rating", y="Compensation", alpha=0.5, color="blue")
plt.vlines(x=3.5,ymin=400000,ymax=1500000,colors="red",alpha=0.5)
plt.hlines(y=900000,xmin=0,xmax=5,colors="red",alpha=0.5)
plt.title("Scatter Plot of Rating vs Compensation")
plt.xlabel("Employee rating out of 5")
plt.ylabel("Compensation")
plt.show()
```

```python
#creating count plot of the rating feature
plt.figure(figsize=(8,4), dpi=100)
sns.countplot(data=df, x="Rating", alpha=0.5, color="blue")
plt.title("Count Plot of Employee Rating")
plt.xlabel("Employee rating out of 5")
plt.ylabel("Count")
plt.show()
```

```python
##creating a histogram of the compensation feature
plt.figure(figsize=(8,4), dpi=100)
sns.histplot(data=df, x="Compensation", bins=20, alpha=0.5, color="blue")
plt.title("Histogram of Compensation")
plt.xlabel("Compensation")
plt.ylabel("Count")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8,4), dpi=100)
sns.histplot(data=df1, x="Age", bins=20, alpha=0.5, color="blue")

# Add vertical lines for median and mean of Age
plt.axvline(df1["Age"].median(), color="red", linestyle="--", linewidth=2, alpha=0.7, label="Median")
plt.axvline(df1["Age"].mean(), color="green", linestyle="-", linewidth=2, alpha=0.7, label="Mean")

plt.title("Histogram of Age")
plt.xlabel("Age")
plt.ylabel("Count")
plt.legend()

plt.show()
```
